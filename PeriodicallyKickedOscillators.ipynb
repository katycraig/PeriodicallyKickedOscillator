Setup

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def psi_tau(theta_0,y_0):
    theta_tau = theta_0 + tau + (sig/lam)*(y_0+A*np.sin(2*np.pi*theta_0))*(1-np.exp(-lam*tau))
    theta_tau = theta_tau %1
    y_tau = np.exp(-lam*tau)*(y_0+A*np.sin(2*np.pi*theta_0))
    return (theta_tau,y_tau)
```

 The limit cycle and its image after one kick

```python
lam = 0.1
A = 0.1
sig = 0.05
tau =1

(theta_0,y_0) = (np.linspace(0,1,200), np.zeros(200))
(theta_tau,y_tau) = psi_tau(theta_0,y_0)

plt.figure(figsize=(8,8))
plt.plot(theta_0,y_0,'.')
plt.plot(theta_tau,y_tau,'.')
```

The effect of increasing shear, sigma = 0.05, 0.25,...
